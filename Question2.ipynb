```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('cybersecurity_attacks.csv')
```

```python
print(data.head())
print(data.info())
```

# Histogram of Attack Types:



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Attack Type')
plt.title('Distribution of Attack Types')
plt.xticks(rotation=90)
plt.show()
```

# Time Series Plot of Attacks:


```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Timestamp', y='Packet Length', hue='Attack Type')
plt.title('Time Series of Attacks')
```

# Bar Plot of Protocols:



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Protocol', order=data['Protocol'].value_counts().index)
plt.title('Distribution of Protocols')
plt.xticks(rotation=90)


```

# Pair Plot for Correlations:



```python
sns.pairplot(data, hue='Attack Type')
```

# Box Plot of Packet Length:



```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Attack Type', y='Packet Length')
plt.title('Packet Length by Attack Type')
plt.xticks(rotation=90)
```

# Violin Plot of Protocol by Packet Length:



```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='Protocol', y='Packet Length')
plt.title('Distribution of Protocol by Packet Length')
plt.xticks(rotation=90)
```

# Heatmap of Correlations:



```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

# Pie Chart of Attack Type Proportions:



```python
attack_type_counts = data['Attack Type'].value_counts()
plt.pie(attack_type_counts, labels=attack_type_counts.index, autopct='%1.1f%%')
```

# Scatter Plot of Packet Length vs. Anomaly Scores:



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Packet Length', y='Anomaly Scores', hue='Attack Type')
plt.title('Packet Length vs. Anomaly Scores')
```

# Box Plot of Attack Type by Packet Length:



```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Attack Type', y='Packet Length')
plt.title('Attack Type by Packet Length')
plt.xticks(rotation=90)
```
